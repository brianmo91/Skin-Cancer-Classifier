Appending new data of melanoma to the original dataset. 
New data consists of the same melanoma images from original dataset and "Augmented" versions of them.
After appending new data, duplicates will also be dropped.

```python
import pandas as pd
```

```python
df = pd.read_csv(r'.\train_original.csv')
print(df.head(10))
```

```python
df['label'].value_counts()
```

```python
df2 = pd.read_csv(r'.\melanoma\melanoma.csv')
print(df2.head(10))
```

```python
df2['label'].value_counts()
```

```python
newdf = df.append(df2)
```

```python
newdf = newdf.drop_duplicates()
```

```python
newdf
```

```python
newdf.to_csv("train.csv",index=False)
```
